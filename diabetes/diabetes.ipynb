```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import core
```

```python
df = np.loadtxt("diabetes.data", delimiter=",")
```

```python
X = df[:, 0:8]
Y = df[:, 8]
```

```python
# create model
model = Sequential()
model.add(Dense(32, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(16, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(4, init='uniform', activation='relu'))
core.Dropout(0.2)
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile Model
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, nb_epoch=2000, batch_size=32)
```

```python
scores = model.evaluate(X, Y)
print ("Accuracy: %.2f%%" %(scores[1]*100))
```

```python
model_json = model.to_json()
with open("diabetesModel.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("diabetesModel.h5")
print("Saved model to disk")
```

```python
import json
from keras.models import model_from_json
json_file = open('diabetesModel.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("diabetesModel.h5")
print("Loaded model from disk")
 
# evaluate loaded model on test data
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
score = loaded_model.evaluate(X, Y, verbose=0)
print("%s: %.2f%%" % (loaded_model.metrics_names[1], score[1]*100))
```
